```python
# 7 countries with lowest numbers of medals in summer
Bottom_7=df4.nsmallest(7,'Summer_Achievement')
x=Bottom_7['Country']
y=Bottom_7['Summer_Achievement']
f1={'font':'serif','size':10,'color':'Black','weight':'bold'}
plt.bar(x,y,color='red')
plt.title('Bottom 7 countries with lowest number of Medals in Summer',f1)
plt.xlabel('Countries',f1)
plt.ylabel('Summer_Achievement',f1)
plt.xticks(rotation=55)
plt.grid(False)
#f1={'font':'serif','size':10,'color':'Black','weight':'bold'}
plt.show()
```

```python
# 7 countries with lowest numbers of medals in winter
Bottom_7=df4.nsmallest(7,'Winter_Achievement')
x=Bottom_7['Country']
y=Bottom_7['Winter_Achievement']
f1={'font':'serif','size':10,'color':'Black','weight':'bold'}
plt.bar(x,y,color='red')
plt.title('Bottom 7 countries with lowest number of Medals in winter',f1)
plt.xlabel('Countries',f1)
plt.ylabel('Winter_Achievement',f1)
plt.xticks(rotation=55)
plt.grid(False)
#f1={'font':'serif','size':10,'color':'Black','weight':'bold'}
plt.show()
# Bottom 7 countries in winter olympics doesnot win any medals
```

```python
# 7 countries with highest numbers of medals in summer
top_7=df4.nlargest(7,'Summer_Achievement')
x=top_7['Country']
y=top_7['Summer_Achievement']
f1={'font':'serif','size':10,'color':'Black','weight':'bold'}
plt.bar(x,y,color='red')
plt.title('Top 7 countries with highest number of Medals in Summer',f1)
plt.xlabel('Countries',f1)
plt.ylabel('Summer_Achievement',f1)
plt.xticks(rotation=55)
plt.grid(False)
#f1={'font':'serif','size':10,'color':'Black','weight':'bold'}
plt.show()
```

```python
# 7 countries with highest numbers of medals in winter
top_7=df4.nlargest(7,'Winter_Achievement')
x=top_7['Country']
y=top_7['Winter_Achievement']
f1={'font':'serif','size':10,'color':'Black','weight':'bold'}
plt.bar(x,y,color='red')
plt.title('Top 7 countries with highest number of Medals in Winter',f1)
plt.xlabel('Countries',f1)
plt.ylabel('Winter_Achievement',f1)
plt.xticks(rotation=55)
plt.grid(False)
#f1={'font':'serif','size':10,'color':'Black','weight':'bold'}
plt.show()
```
